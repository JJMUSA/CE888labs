```python
import pandas
import numpy as np
import pandas as pd

#from __future__ import print_function

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


import matplotlib.pyplot as plt
import matplotlib.cm as cm

%matplotlib inline  
import seaborn as sns
```

```python
df_demo = pd.DataFrame()
df = pd.read_csv("hUSCensus1990raw50K.csv.bz2",compression = "bz2")

df_demo["AGE"] = df[["AGE"]].copy()
df_demo["INCOME"] = df[["INCOME" + str(i) for i in range(1,8)]].sum(axis = 1)

df_demo["YEARSCH"] = df[["YEARSCH"]].copy()
df_demo["ENGLISH"] = df[["ENGLISH"]].copy()
df_demo["FERTIL"] = df[["FERTIL"]].copy()
df_demo["YRSSERV"] = df[["YRSSERV"]].copy()



df_demo = pd.get_dummies(df_demo, columns = ["ENGLISH", "FERTIL" ])
print(df_demo.columns)
```

```python
#pariplot of data

sns.pairplot(data=df_demo)
plt.show()
```

```python
#looping through cluster_size of 2-10 and picking the one with the one with lowest silhouette s=core
X = df_demo.values[np.random.choice(df_demo.values.shape[0], 10000)]

from sklearn import metrics
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_db = sc.fit_transform(X)


sil_score=1
lowest_cluster=2
for i in range(lowest_cluster,11):
    labels = KMeans(n_clusters = i).fit_predict(X_db)
    sl=metrics.silhouette_score(X_db, labels)
    print(sl)
    if sil_score>sl:
        sil_score=sl
        lowest_cluster=i
print("cluster %d has the lowest sil_score with sil_score of:%0.3f"%(lowest_cluster,sil_score))
    
```

```python
# getting 10 different silhouette_score for cluster_size of 2-11
sil_confidence=np.empty((10,10),dtype=np.float64)
for i in range(2,11):
    sil=[]
    for j in range(10):
        labels=KMeans(n_clusters=i).fit_predict(X_db)
        sil.append(metrics.silhouette_score(X_db,labels))
    sil=np.asarray(sil)
    print(sil.shape)
    sil_confidence[i-2] =sil
sil_confidence
```

```python
# plotting a seaborn point plot of silhouette_scores
df=pd.DataFrame(sil_confidence)
rows=list(range(0,sil_confidence.shape[0]))
col=[list(df.columns)]
print(rows)
print(col)

sns.pointplot(data=df,join=False)
```

```python
# measuring the silhouette of 10 AgglomerativeClustering for cluster_size of 2-10
from sklearn.cluster import AgglomerativeClustering
sillhouette_score=np.empty([10,10])
for i in range(2,11):
    sil=[]
    for j in range(0,10):
        labels=AgglomerativeClustering(n_clusters=i).fit_predict(X_db)
        sil.append(metrics.silhouette_score(X_db,labels))
    sillhouette_score+=sil
print(sillhouette_score)
sns.pointplot(data=sillhouette_score,join=False)
```

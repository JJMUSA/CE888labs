```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
data=pd.read_csv("jester-data-1.csv",header=None)
data=data.drop([0],axis=1)
data.head()
```

```python
#split data into validation/ training set
Partialdata,val=train_test_split(data,test_size=0.10,random_state=0)
val.head()
```

```python
#testSet= validation set with all rating set to 99.0
testSet=pd.DataFrame(val)
testSet=testSet[testSet!=99.0]
testSet=testSet.dropna()
print(testSet.head())
valSet=testSet.applymap(lambda x: 99)
valSet.head
```

```python
#generating random features for jokes
item_features=np.random.random((valSet.shape[1],2))
print(item_features.shape)
item_features.shape
valSet=valSet.values
print(valSet)
```

```python
#generating random prefrence values for users
user_prefrences=np.random.random((valSet.shape[0],2))
print(user_prefrences)


```

```python

#training
def predict_rating(user_id,item_id):
    
    user_preference= user_prefrences[user_id]
    #print("User Preference: ",user_preference)
    item_preference = item_features[item_id]
    
    #print("Prediction: " , user_preference.dot(item_preference))
    return user_preference.dot(item_preference)


def train(user_id, item_id, rating,alpha = 0.0001):
    
    #print item_id
    prediction_rating = predict_rating(user_id, item_id)
    err = alpha * ( prediction_rating- rating );
    #print err
    user_pref_values = user_prefrences[user_id][:]
    user_prefrences[user_id] -= err * item_features[item_id]
    item_features[item_id] -= err * user_pref_values
    return (err)
    


def sgd(iterations = 1000):
    """ Iterate over all users and all items and train for 
        a certain number of iterations
    """
    for iteration in range(0,iterations):
        error = []
        for user_id in range(0,user_prefrences.shape[0]):
            for item_id in range(0,item_features.shape[0]):
                
                rating = testSet[user_id][item_id]
                #print(rating)
                if(rating!=99.0):
                    err = train(user_id,item_id,rating)
                    error.append(err)
        mse = (np.array(error) ** 2).mean()          
        print (mse)
sgd()
```

```python
#making predicitions on valSet
MSE=0
for user in range(valSet.shape[0]):
    for joke in range(valSet.shape[1]):
        prediction=predict_rating(user,joke)
        valSet[user][joke]=prediction
        if testSet[user][joke]!=0:
            MSE+=((testSet[user][joke]-prediction)/testSet[user][joke])**2
MSE/=valSet.shape[0]
print("MSE = ",MSE)
```

```python
#puting true values and predictions in a table to make comaprison
comp=[zip(testSet[i],valSet[i]) for i in range(valSet.shape[0])]
compTable=pd.DataFrame(comp)
compTable.applymap(lambda x: "(%2.3f|%2.3f)"%(x[0],x[1]))
```
